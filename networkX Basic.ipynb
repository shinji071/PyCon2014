##NetworkXによるグラフ理論
networkxはPython上で稼働する無料のグラフ生成・分析ライブラリです

###1. networkxの特徴
* 無向、有向グラフをPythonのデータ構造で記述可能
* ノード、エッジには任意の属性を持たせる事が可能
* ネットワーク分析・経路探索などの各種メソッドを提供
* Open Source : BSD License

###2. 概要
####Graph
* グラフを形成するノード、エッジを管理するクラス
* ノード、エッジの位置関係、属性などを編集可能
* 経路探索といったメソッドも持つ

####グラフ種類
* G = Graph() 閉路を含まない無向グラフ
* D = DiGraph() 閉路を含まない有効グラフ
* G = MultiGraph() 多重枝を許した有効グラフ
* D = MultiDiGraph() 多重枝を許した有効グラフ

* ※グラフへのノード・エッジの追加・削除は以下の例を参考にして下さい。

### 空のグラフ作成

```python
from __future__ import print_function
import networkx as nx
G=nx.Graph()
```

### ノード・エッジの追加

```python
#グラフにノードを1つ追加
G.add_node(1)
#ノード名は、任意の英数字が可能。リストで渡す事も可能。
G.add_nodes_from([100,"abc"])

#グラフに（ノード1と100を繋ぐ)エッジを1つ追加
G.add_edge(1,100)
#定義されていないノードと繋ぐエッジも追加可能（ノードは自動的に追加される）
#リストで渡す事も可能
G.add_edges_from([(100,"abc"),(1,"abc"), (100,3)])

#G.nodes(), G.edges()関数で、追加済のノード、エッジリストを取得可能
print("NODES:",G.nodes(), "\nEDGES:",G.edges())
#for key, item in G.node.items():
#    print key, item
nx.draw(G)
```

### ノード・エッジへの属性情報付加

```python
#ノード追加時に属性値を与える(priceという属性を作り、100円とした)
G.add_node(38, price=100)
#既存ノードに属性値を与える
G.node[1]["price"] = 300

#エッジ追加時に属性値を与える（weightという属性を作り、4.7とした）
G.add_edge(1,2,weight=4.7)
#リスト追加時にも、属性値を与える事が可能
G.add_edges_from([(1,3),(3,"abc")], color='red')
#リスト追加時に、別個の属性値を与える事も可能
G.add_edges_from([(11,12,{'color':'blue'}), (12,13,{'weight':8})])
#既存エッジに属性値を与える
G.edge[11][12]['weight'] = 100

#任意の属性を設定可能であるが、多くの場合、weight属性が重要となる
#add_weighted_edges_fromは、(node1,node2,value)のタプルを要素としたリストを与える事により、
#自動で対応エッジにweight属性値を設定する
G.add_weighted_edges_from([(1,3,100),(1,'abc',200)])

print("NODES:",G.nodes(data=True), "\nEDGES:",G.edges(data=True))
G.clear()
```

### グラフのグラフへの追加

```python
H = nx.path_graph(10)
G=nx.Graph()
G.add_nodes_from(H)
print(G.nodes(), G.edges())
G.add_edges_from(H.edges())
print(G.nodes(), G.edges())
nx.draw(G)
```

### ノード・リンクの削除

```python
print("削除前")
print("NODES:",G.nodes(), "\nEDGES:",G.edges())
print("------------------------------------")
#ノードを削除する
print("ノード7, 8を削除。対応するエッジも削除される")
G.remove_nodes_from([7,8])
print("NODES:",G.nodes(), "\nEDGES:",G.edges())
print("------------------------------------")
#エッジを削除する
print("エッジ1-3, 3-5を削除。")
G.remove_edges_from([(1,3),(3,5)])
print("NODES:",G.nodes(), "\nEDGES:",G.edges())
print("------------------------------------")
#グラフを削除する
print("グラフ内の全ノード、アークを削除")
G.clear()
print("NODES:",G.nodes(), "\nEDGES:",G.edges())
```

### 有向グラフの生成

```python
#空の有向グラフを生成する
H=nx.DiGraph()
edgelist=[(0,1),(1,2),(2,3)]
#上記のようなエッジは、0->1, 1->2, 2->3のエッジとして認識される
H.add_edges_from(edgelist)
print(H.nodes(), H.edges())
#neighbors(node)はnodeから到達可能な最近接ノードを返す
print(H.neighbors(1))
nx.draw(H)
```

## アルゴリズム

#### 深さ優先探索

```python
ladder = nx.ladder_graph(5)
ladder.add_edge(0,10)
nx.draw(ladder)
print("edges ordered by DFS")
for e in nx.dfs_edges(ladder,1):
    print(e, end='')
print("\ntrees ordered by DFS")
for tree in nx.dfs_tree(ladder,1):
    print(tree, end='')
print("\n探索の前後関係をDictionaryで返す")
print(nx.dfs_predecessors(ladder,1))
print(nx.dfs_successors(ladder,1))

print("Preorder順のノード(stackに入る順)")
for n in nx.dfs_preorder_nodes(ladder,1):
    print(n, end='')
print("\nPostorder順のノード(stackから取り出す順)")
for n in nx.dfs_postorder_nodes(ladder,1):
    print(n, end='')
```

#### 幅優先探索

```python
ladder = nx.ladder_graph(5)
ladder.add_edge(0,10)
nx.draw(ladder)
print("edges ordered by BFS")
for e in nx.bfs_edges(ladder,1):
    print(e, end='')
print("\ntrees ordered by DFS")
for tree in nx.bfs_tree(ladder,1):
    print(tree, end='')
print("\n探索の前後関係をDictionaryで返す")
print(nx.dfs_predecessors(ladder,1))
print(nx.dfs_successors(ladder,1))
```

#### サイクル検知

```python
def find_all_cycles(G, source=None, cycle_length_limit=None):
    """forked from networkx dfs_edges function. Assumes nodes are integers, or at least
    types which work with min() and > ."""
    if source is None:
        # produce edges for all components
        nodes=[i[0] for i in nx.connected_components(G)]
    else:
        # produce edges for components with source
        nodes=[source]
    # extra variables for cycle detection:
    cycle_stack = []
    output_cycles = set()
    
    def get_hashable_cycle(cycle):
        """cycle as a tuple in a deterministic order."""
        m = min(cycle)
        mi = cycle.index(m)
        mi_plus_1 = mi + 1 if mi < len(cycle) - 1 else 0
        if cycle[mi-1] > cycle[mi_plus_1]:
            result = cycle[mi:] + cycle[:mi]
        else:
            result = list(reversed(cycle[:mi_plus_1])) + list(reversed(cycle[mi_plus_1:]))
        return tuple(result)
    
    for start in nodes:
        if start in cycle_stack:
            continue
        cycle_stack.append(start)
        
        stack = [(start,iter(G[start]))]
        while stack:
            parent,children = stack[-1]
            try:
                child = next(children)
                
                if child not in cycle_stack:
                    cycle_stack.append(child)
                    stack.append((child,iter(G[child])))
                else:
                    i = cycle_stack.index(child)
                    if i < len(cycle_stack) - 2: 
                      output_cycles.add(get_hashable_cycle(cycle_stack[i:]))
                
            except StopIteration:
                stack.pop()
                cycle_stack.pop()
    
    return [list(i) for i in output_cycles]
```

### Networkを用いた簡単なシミュレーション (MetaPopulation)

[参考サイト]http://timotheepoisot.fr/2012/05/18/networkx-metapopulations-python/

```python
import numpy as np
nodes = 100   # ノード数
p_ext = 0.01  # 絶滅確率
p_col = 0.014 #居住確率
p_init = 0.1 #初期居住確率
distance = 1.2 #アークで接続するノード間距離

class house:
    def __init__(self,status=0,pos=(0,0)):
        self.status = status
        self.pos = pos
    def __str__(self):
        return(str(self.status))
```

```python
G = nx.Graph()
for i in range(nodes):
    stat = 1 if np.random.uniform() < p_init else 0
    pos  = (np.random.uniform()*10-5,np.random.uniform()*10-5)
    G.add_node(house(stat,pos))

for p1 in G.nodes():
    for p2 in G.nodes():
        dist = np.sqrt((p1.pos[1]-p2.pos[1])**2+(p1.pos[0]-p2.pos[0])**2)
        if dist <= distance:
            G.add_edge(p1,p2)
            
occup = [n.status for n in G]
pos = {}
for n in G.nodes():
    pos[n] = n.pos
nx.draw(G,node_color=occup,with_labels=False,cmap=plt.cm.Greys,pos=pos,vmin=0,vmax=1)
```

```python
time = [0]
#居住率
occupancy = [np.sum([n.status for n in G])/float(nodes)]
for timestep in range(8000):
    # 絶滅判定
    for n in G.nodes():
        if (n.status == 1 and np.random.uniform() < p_ext):
            n.status = 0
    # 侵攻判定
    for n in G.nodes():
        if n.status == 1:
            neighb = G[n] # = nx.neighbors(G)
            for nei in neighb:
                if nei.status == 0:
                    if np.random.uniform() < p_col:
                        nei.status = 1
                        break
    time.append(timestep+1)
    occupancy.append(np.sum([n.status for n in G])/float(nodes))
```
